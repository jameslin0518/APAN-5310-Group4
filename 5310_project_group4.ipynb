```python
import psycopg

def get_connection():
    """Establishes a connection to the PostgreSQL database."""
    try:
        return psycopg.connect(
            host="project.cpwow42uutyx.us-east-2.rds.amazonaws.com",
            port='5432',
            dbname="project",
            user="postgres",
            password="5310project"
        )
    except Exception as e:
        print(f"Failed to connect to the database: {e}")
        return None

def clear_tables(conn):
    """Drops all existing data in the tables to avoid duplicate key errors."""
    try:
        with conn.cursor() as cur:
            cur.execute("TRUNCATE TABLE PurchaseOrderItems, PurchaseOrders, Inventory, WorkHours, StoreExpenses, Products, Categories, Vendors CASCADE;")
            conn.commit()
            print("All tables cleared successfully.")
    except Exception as e:
        print(f"An error occurred while clearing tables: {e}")

def insert_categories(conn, df_sql_categories):
    """Inserts data into the Categories table."""
    try:
        with conn.cursor() as cur:
            for index, row in df_sql_categories.iterrows():
                cur.execute("""
                    INSERT INTO Categories (CategoryID, CategoryName)
                    VALUES (%s, %s)
                """, (row['CategoryID'], row['CategoryName']))
            conn.commit()
            print("Categories inserted successfully.")
    except Exception as e:
        conn.rollback()
        print(f"An error occurred while inserting Categories: {e}")

def insert_products(conn, df_sql_products):
    """Inserts data into the Products table."""
    try:
        with conn.cursor() as cur:
            for index, row in df_sql_products.iterrows():
                cur.execute("""
                    INSERT INTO Products (ProductID, ProductName, CategoryID, SuggestedPrice, Description)
                    VALUES (%s, %s, %s, %s, %s)
                """, (row['ProductID'], row['ProductName'], row['CategoryID'], row['SuggestedPrice'], row['Description']))
            conn.commit()
            print("Products inserted successfully.")
    except Exception as e:
        conn.rollback()
        print(f"An error occurred while inserting Products: {e}")

def insert_vendors(conn, df_sql_vendors):
    """Inserts data into the Vendors table."""
    try:
        with conn.cursor() as cur:
            for index, row in df_sql_vendors.iterrows():
                cur.execute("""
                    INSERT INTO Vendors (VendorID, VendorName, ContactName, ContactPhone, Email, Address)
                    VALUES (%s, %s, %s, %s, %s, %s)
                """, (row['VendorID'], row['VendorName'], row['ContactName'], row['ContactPhone'], row['Email'], row['Address']))
            conn.commit()
            print("Vendors inserted successfully.")
    except Exception as e:
        conn.rollback()
        print(f"An error occurred while inserting Vendors: {e}")

def insert_purchase_orders(conn, df_sql_PurchaseOrders):
    """Inserts data into the PurchaseOrders table and return a set of all OrderIDs."""
    valid_order_ids = set()
    try:
        with conn.cursor() as cur:
            for index, row in df_sql_PurchaseOrders.iterrows():
                status = row['Status']
                if status not in ['Delivered', 'Cancelled', 'Ordered']:
                    print(f"Invalid status {status} for OrderID {row['OrderID']}, skipping insertion.")
                    continue

                # Insert all orders, including Cancelled
                cur.execute("""
                    INSERT INTO PurchaseOrders (OrderID, VendorID, OrderDate, ExpectedDeliveryDate, ActualDeliveryDate, Status)
                    VALUES (%s, %s, %s, %s, %s, %s)
                """, (row['OrderID'], row['VendorID'], row['OrderDate'], row['ExpectedDeliveryDate'], row['ActualDeliveryDate'], status))
                valid_order_ids.add(row['OrderID'])

            conn.commit()
            print("Purchase Orders inserted successfully.")
    except Exception as e:
        conn.rollback()
        print(f"An error occurred while inserting PurchaseOrders: {e}")
    
    return valid_order_ids

def insert_purchase_order_items(conn, df_sql_PurchaseOrderItems, valid_order_ids):
    """Inserts data into the PurchaseOrderItems table only for valid OrderIDs."""
    try:
        with conn.cursor() as cur:
            for index, row in df_sql_PurchaseOrderItems.iterrows():
                if row['OrderID'] not in valid_order_ids:
                    print(f"Skipping PurchaseOrderItem with OrderID {row['OrderID']} as it is not in the valid list.")
                    continue

                cur.execute("""
                    INSERT INTO PurchaseOrderItems (OrderItemID, OrderID, ProductID, Quantity, UnitPrice)
                    VALUES (%s, %s, %s, %s, %s)
                """, (row['OrderItemID'], row['OrderID'], row['ProductID'], row['Quantity'], row['UnitPrice']))
            conn.commit()
            print("Purchase Order Items inserted successfully.")
    except Exception as e:
        conn.rollback()
        print(f"An error occurred while inserting PurchaseOrderItems: {e}")

def insert_inventory(conn, df_sql_inventory):
    """Inserts data into the Inventory table."""
    try:
        with conn.cursor() as cur:
            for index, row in df_sql_inventory.iterrows():
                cur.execute("""
                    INSERT INTO Inventory (InventoryID, StoreID, ProductID, Quantity, LastRestocked, ReorderPoint)
                    VALUES (%s, %s, %s, %s, %s, %s)
                """, (row['InventoryID'], row['StoreID'], row['ProductID'], row['Quantity'], row['LastRestocked'], row['ReorderPoint']))
            conn.commit()
            print("Inventory inserted successfully.")
    except Exception as e:
        conn.rollback()
        print(f"An error occurred while inserting Inventory: {e}")

def insert_store_expenses(conn, df_sql_StoreExpenses):
    """Inserts data into the StoreExpenses table."""
    try:
        with conn.cursor() as cur:
            for index, row in df_sql_StoreExpenses.iterrows():
                # Correct date format if needed
                month = row['Month']
                if isinstance(month, str) and len(month) == 7:
                    month += '-01'  # Convert "YYYY-MM" to "YYYY-MM-DD"

                cur.execute("""
                    INSERT INTO StoreExpenses (StoreID, Month, Utility, Marketing, Security, Supplies, Maintenance, OtherExpense)
                    VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
                """, (row['StoreID'], month, row['Utility'], row['Marketing'], row['Security'], row['Supplies'], row['Maintenance'], row['OtherExpense']))
            conn.commit()
            print("Store Expenses inserted successfully.")
    except Exception as e:
        conn.rollback()
        print(f"An error occurred while inserting StoreExpenses: {e}")

def insert_workhours(conn, df_sql_workhours):
    """Inserts data into the WorkHours table."""
    try:
        with conn.cursor() as cur:
            for index, row in df_sql_workhours.iterrows():
                cur.execute("""
                    INSERT INTO WorkHours (WorkHoursID, EmployeeID, StoreID, StartTime, EndTime)
                    VALUES (%s, %s, %s, %s, %s)
                """, (row['WorkHoursID'], row['EmployeeID'], row['StoreID'], row['StartTime'], row['EndTime']))
            conn.commit()
            print("Work Hours inserted successfully.")
    except Exception as e:
        conn.rollback()
        print(f"An error occurred while inserting WorkHours: {e}")

def run_inserts():
    """Runs the data insertion for all tables."""
    conn = get_connection()
    
    if conn is None:
        print("Unable to establish a connection.")
        return

    try:
        # Clear existing data
        clear_tables(conn)

        # Insert data into tables
        insert_categories(conn, df_sql_categories)
        insert_products(conn, df_sql_products)
        insert_vendors(conn, df_sql_vendors)

        # Insert PurchaseOrders and get all OrderIDs
        all_order_ids = insert_purchase_orders(conn, df_sql_PurchaseOrders)
        
        # Insert PurchaseOrderItems with valid OrderIDs
        insert_purchase_order_items(conn, df_sql_PurchaseOrderItems, all_order_ids)

        # Insert remaining data
        insert_inventory(conn, df_sql_inventory)
        insert_store_expenses(conn, df_sql_StoreExpenses)
        insert_workhours(conn, df_sql_workhours)
    except Exception as e:
        print(f"An error occurred during data insertion: {e}")
    finally:
        conn.close()
        print("Database connection closed.")

# Run the insert operations
run_inserts()
```
